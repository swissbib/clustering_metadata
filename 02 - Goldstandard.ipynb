## Transform list columns to string columns

```python
def transform_list_to_string(df, column):
    df[column] = [', '.join(map(str, list_element)) for list_element in df[column]]
    
    return df
```

```python
list_columns = ['coordinate', 'corporate', 'pages', 'part', 'pubword',
                'ttlfull', 'ttlpart']
array_of_strings_columns = ['person']

for list_col in list_columns:
    df_m = transform_list_to_string(df_m, list_col)
    df_s = transform_list_to_string(df_s, list_col)
    df_u = transform_list_to_string(df_u, list_col)

for aos_col in array_of_strings_columns:
    df_m = transform_list_to_string(df_m, aos_col)
    df_s = transform_list_to_string(df_s, aos_col)
    df_u = transform_list_to_string(df_u, aos_col)
    
# Additional
df_s = transform_list_to_string(df_s, '035liste')
df_u = transform_list_to_string(df_u, '035liste')
```

```python
# Python RegEx library for string compare
import re

def master_slaves_compare(masters, slaves, i, j):
    search_string = re.sub(r'\(', '\(', masters['035liste'].loc[i][j])
    search_string = re.sub(r'\)', '\)', search_string)
    print('\nmaster index {:d}, slave relation {:d}, {:s}'.format(i, j, search_string))
    
    slave_index = slaves[slaves['035liste'].str.contains(search_string)].index
    
    return slave_index
```

```python
for i in range(len(df_m)):
    for j in range(len(df_m['035liste'].loc[i])):
        slave_index = master_slaves_compare(df_m, df_u, i, j)

        df_united = pd.concat([df_m.loc[[i]], df_u.iloc[slave_index]], axis=0)
        print(df_united.drop(columns=['035liste', 'docid']).duplicated())
```

## Find duplicates

```python
def find_duplicates(df):
    df_duplicates = df[df.drop(columns=['035liste', 'docid']).duplicated(keep=False)].sort_values(by='ttlfull')
    
    return df_duplicates
```
